```python
%load_ext autoreload
%autoreload 2

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import mra
from mra import *
import models
from models import *

os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
```

```python
def make_dataset(X, Y, f):
    x = tf.data.Dataset.from_tensor_slices(X)
    y = tf.data.Dataset.from_tensor_slices(Y)
    x = x.map(f)
    xy = tf.data.Dataset.zip((x, y))
    xy = xy.shuffle(10000)
    return xy

df = pd.read_excel('../datasets/credit-card.xls', header=1, index_col=0).sample(frac=1)
x = df.drop(columns=["default payment next month"]).to_numpy()
x = (x - np.min(x, axis=0))/(np.max(x, axis=0) - np.min(x, axis=0))
y = df["default payment next month"].to_numpy().reshape((len(x), 1)).astype("float32")
train_ds = make_dataset(x, y, lambda t: t)
```

```python
main = mra(train_ds)
make_f, make_g, make_generator = models.make_mlp(
        attr_num=23,
        class_num=2,
        split=4,
        fg_units=64,
        fg_act="relu",
        generator_units=[512,256,128,64]
)
main.train(
    make_f,
    make_g,
    loss_fn=tf.keras.losses.BinaryCrossentropy(),
    batch_size=32,
    epoch=5,
    lr=0.001,
    verbose=True, log_every=1000
)
```

```python
log_attack = {}

log_attack['B'] = main.attack(
    attack_iter=100,
    make_generator=make_generator,
    input_noise=False,
    input_z=True,
    model_leak=True,
    lr_x=0.0001,
    lr_f=0.00001,
    epoch=50,
    iter_x=40,
    iter_f=1,
    verbose=True, log_every=1
)
```

```python
log_attack['C'] = main.attack(
    attack_iter=100,
    make_generator=make_generator,
    input_noise="normal",
    input_z=False,
    model_leak=True,
    lr_x=0.0001,
    lr_f=0.00001,
    epoch=50,
    iter_x=40,
    iter_f=1,
    verbose=True, log_every=1
)
```

```python
log_attack['D'] = main.attack(
    attack_iter=100,
    make_generator=None,
    input_noise=False,
    input_z=False,
    model_leak=True,
    lr_x=0.0001,
    lr_f=0.00001,
    epoch=50,
    iter_x=40,
    iter_f=1,
    verbose=True, log_every=1
)
```

```python
log_attack['E'] = main.attack(
    attack_iter=100,
    make_generator=make_generator,
    input_noise="normal",
    input_z=True,
    model_leak=False,
    lr_x=0.0001,
    lr_f=0.0001,
    epoch=50,
    iter_x=40,
    iter_f=40,
    verbose=True, log_every=1
)
```

```python
log_attack['F'] = main.attack(
    attack_iter=100,
    make_generator=None,
    input_noise=False,
    input_z=False,
    model_leak=False,
    lr_x=0.0001,
    lr_f=0.0001,
    epoch=50,
    iter_x=40,
    iter_f=40,
    verbose=True, log_every=1
)
```

```python
import pickle
with open('./log/ablation.pkl', 'wb') as f:
    pickle.dump(log_attack, f)
```

```python
import matplotlib.pyplot as plt

with open('./log/credit.pkl', 'rb') as f:
    log_attack['A'] = pickle.load(f)[4]

from matplotlib.pyplot import figure

figure(figsize=(8, 6))
# plt.gca().set_yscale("logit")
plt.xlim([1,100])
plt.ylim([0,0.35])
plt.grid()
plt.ylabel('MSE per feature')
plt.xlabel('Batch distance from the obtained model snapshot')
ran = [i+1 for i in range(100)]
plt.plot(
    ran, log_attack['B'][:,0], 'k--',
    ran, log_attack['B'][:,1], 'k:',
    ran, log_attack['A'][:,2], 'gD-',
    ran, log_attack['B'][:,2], 'r^-',
    ran, log_attack['C'][:,2], 'bs-',
    ran, log_attack['D'][:,2], 'cv-',
    ran, log_attack['E'][:,2], 'mx-',
    ran, log_attack['F'][:,2], 'yo-',
    markevery=np.arange(0, 101, 9))
plt.gca().fill_between(ran, np.minimum(log_attack['B'][:,0],log_attack['B'][:,1]), 0.35, facecolor='red',alpha=0.05)
plt.legend(['RGU', 'RGN', 'A', 'B', 'C', 'D', 'E', 'F'], loc='upper center', ncol=4)
plt.savefig("fig/ablation.pdf", bbox_inches='tight')
```
